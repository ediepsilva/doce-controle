## Testes Funcionais

```python
# Teste 2: Validar sintaxe PHP dos arquivos principais

import subprocess

print('═' * 60)
print('TESTE 2: Validação de Sintaxe PHP')
print('═' * 60)

arquivos_php = [
    'config.php',
    'index.php',
    'clientes.php',
    'receitas.php',
    'estoque.php',
    'pedidos.php'
]

erros_encontrados = []

for arquivo in arquivos_php:
    caminho = os.path.join(projeto_dir, arquivo)
    try:
        resultado = subprocess.run(
            ['php', '-l', caminho],
            capture_output=True,
            text=True,
            timeout=5
        )
        if 'No syntax errors' in resultado.stdout or resultado.returncode == 0:
            print(f'✓ {arquivo} - Sintaxe OK')
        else:
            print(f'✗ {arquivo} - Erro encontrado')
            erros_encontrados.append({
                'arquivo': arquivo,
                'erro': resultado.stdout + resultado.stderr
            })
    except Exception as e:
        print(f'⚠ {arquivo} - Não foi possível validar: {str(e)}')

if erros_encontrados:
    print('\n⚠ Erros de sintaxe encontrados:')
    for erro in erros_encontrados:
        print(f'\n{erro["arquivo"]}:')
        print(erro['erro'])
else:
    print('\n✓ Todos os arquivos PHP têm sintaxe válida!\n')
```
